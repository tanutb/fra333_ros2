The principle of robotics heaviliy ties to the idea of changing variables from one space to the other. Often times, a user indicates the task of the robot by giving some information in the **TASKSPACE**. However, the robot is actuated in its own **CONFIGURATION SPACE**. For example, if the user wants a 3-DOF RRR robotics arm to move its end-effector to a particular position in 3D space, the robot has to move all 3 joints accordingly. Surely, each revolute joint cannot independently correspond to each componenet of the position. Some means of conversion are required to map the motion of the joints to the motion of the end-effector. If the mapping of the motion is regardless of force nor the effort of the joints, we call this "**KINEMATICS**". There are several ways to categorize kinematics. We can categorize kinematics based on the quantities of interest, i.e. position kinematics, orientation kinematics, or velocity kinematics. 

**POSITION KINEMATICS** is considered to be the most fundamental kinematics to study since most of other kinematics can be derived from the position kinematics. In position kinematics, the **TASKSPACE** refers to a set of all possible position of the robot's end-effector relative and with respect to the global coordinate frame, and the **CONFIGURATION SPACE** refers to a set of all possible joint configuration. The mapping of a robot's joint configuration to a position of its end-effector is known as "**FORWARD KINEMATICS**", which can be generallly represented by the following equation.

$$
\pvec{p}{G}{e}{G} = \mathbf{fk}(\mathbf{q})
$$

   where $\mathbf{q}$ is the given joint configuration and $\pvec{p}{G}{e}{G}$ is the corresponding position of the end effector

In other words, given the joint configuration of a robot, forward (position) kinematics solves for the corresponding position of the end-effector.

In the other hand, the mapping of a position of the robot's end-effector to the joint configuration is known as "**INVERSE KINEMATICS**", which can be generallly represented by the following equation.

$$
\mathbf{q} = \mathbf{ik}(\pvec{p}{G}{e}{G})
$$

Although there is only one position that can be mapped from a set of joint configuration, there are multiple joint configurations that can result to the same end-effector position. Therefore, a new set of parameters called "**ARM CONFIGURATION**", denoted by $\mathbf{\gamma}\in{\{-1,1\}}^m$, can be used to specify the variation of the inverse kinematics solution.

$$
\mathbf{q} = \mathbf{ik}(\pvec{p}{G}{e}{G},\mathbf{\gamma})
$$

<font size=5><B>Summary on Robot Visualization Pipeline in ROS2:</B></font>

Being one of readily available tools in ROS2, *rviz2* is a tool that allows the user to *VISUALIZE* information (topics) in ROS2 network from various points of view. One of many topics that *rviz2* can visualize is "robot_description", which is considered to be one of the most unique topics in ROS2. Although being a std_msgs/String, The robot_description cannot just be anything. It must follow a particular convention in an XML format, otherwise *rviz2* will not display properly. In theory, one can write an XML-formatted string and publish it to *rviz2*, but ROS2 has better tool for this workflow. 

Displaying a robot also requires another special topic known as a "Transformation Tree". A transformation tree is implemented as a backbone for robot's forward kinematics that keeps track of all specified coordinate frames in term of their position, orientation, and time stamp when they were published. To increase its efficiency, ROS2 implements categorized trabsformation tree into 2 separate yet connected topics known as */tf_static* and */tf*. Any "fixed" transformation between coordinate frames will be described in */tf_static*. If the transformation between frames can change over time, it can be describe in */tf*. *rviz2* will use these 2 topics and combine them to crete a complete transformation tree. Sometimes, the users develop their own nodes, which broadcast some transfromation to the transformation tree manually. Technically, one can definitely write a complete transformation tree for a manipulator, but ROS2 also has a convenient tool for broadcasting transformation treee automatically based on the given "robot_description".  

Writing a pure XML file describing a robot's appearance and broadcasting its transformation tree manually can be cumbersome tasks. Luckily, there is a node known as *robot_state_publisher*. This node allows users to parse in the robot's configuration, known as "/joint_states", and a URDF file to the node and publish both transformation tree and the "robot_description".
