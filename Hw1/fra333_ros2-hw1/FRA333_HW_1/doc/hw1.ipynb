<font size=5><B>Summary :</B></font>

$$
\def\pvec#1#2#3#4{\mathbf{#1}_{\mathrm{#2},\mathrm{#3}}^{\mathrm{#4}}}

ท่าทาง (Pose) คือลักษณะของวัตถุในปริภูมิ 3 มิติซึ่งประกอบไปด้วยทิศทางหมุน (Orientation) และ ตำแหน่ง (Position) เราสามารถระบุตำแหน่งของจุด  ที่สัมพัทธ์กับจุดกำเนิดของพิกัดเฟรม  และอ้างอิงกับพิกัดเฟรม โดยใช้เวกเตอร์(หลัก) 3 มิติ 
$$
\pvec pGAG
 = \begin{bmatrix} p_{\mathrm{x}} & p_{\mathrm{y}} & p_{\mathrm{z}} \end{bmatrix}^\top
$$


แต่ทิศทางหมุนสามารถระบุได้หลากหลายรูปแบบ เช่น ZYZ-Euler angles 

$$
\rvec r{ZYZ}G = \begin{bmatrix} \phi & \theta & \psi \end{bmatrix}^\top
$$

หรือ Roll-Pitch-Yaw angles

$$\rvec r{RPY}G =\begin{bmatrix} r_{\mathrm{x}} & r_{\mathrm{y}} & r_{\mathrm{z}} \end{bmatrix}^{\top }$$


นอกจากนี้ทิศทางหมุนยังสามารถเขียนในรูปของเมตริกซ์ตั้งฉากพิเศษ (Special Orthogonal Matrix ${\mathbf{SO}}_3$) หรือเมตริกซ์การหมุนได้อีกด้วย  (Rotation Matrix)

$$
\rotm RAG = \begin{bmatrix} \uvec xAG  & \uvec yAG  & \uvec zAG  \end{bmatrix}^\top
$$

ดังนั้น เราไม่สามารถใช้เมตริกซ์ 3คูณ3ใดๆได้เลย เมตริกซ์การหมุนนั้นจำเป็นต้องมีคุณสมบัติดังต่อไปนี้

$$
\rotm RAG {\rotm RAG}^\top = \mathbf{I_{3}}
$$

$$
\mathrm{det}(\rotm RAG)=1
$$

นอกจากการอธิบายท่าทางของวัตถุในปริภูมิ 3 มิติแล้ว เรายังสามารถใช้หลักการคำนวณเดียวกันในการแปลง (transform) วัตถุได้อีกด้วย ใช่ว่าการแปลงนั้นจะสามารถใช้ได้กับวัตถุทุกชนิด เราจำเป็นที่จะต้องคำนึงถึงประเภทของวัตถุนั้นอีกด้วย

หากเรามีทิศทางหมุน $\rotm RBG$ และทำการหมุนรอบแกน"ตรึง"ด้วยการหมุน $\mathbf{R}$ เราจะได้ทิศทางหมุนอันใหม่เป็นดังต่อไปนี้
$${\mathbf{R}}^{\prime } =\mathbf{R}\rotm RBG$$
หากเรามีตำแหน่ง  และทำการหมุนรอบแกน"ปัจจุบัน"ด้วยการหมุน แล้วทำการเลื่อนด้วยระยะ  เราจะได้ตำแหน่งใหม่เป็นดังต่อไปนี้
$${\mathbf{p}}^{\prime } =\mathbf{R}\pvec pGBG +\mathbf{p}$$
เราสามารถเขียนสมการเบื้องต้นในรูปของเมตริกซ์การแปลงแบบเอกพันธุ์ (Homogenous Transformation Matrix) ได้ดังต่อไปนี้
$$\left\lbrack \begin{array}{c}
{\mathbf{p}}^{\prime } \\
\cdots \\
1
\end{array}\right\rbrack =\mathbf{H}\left\lbrack \begin{array}{c}
\pvec pGBG \\
\cdots \\
1
\end{array}\right\rbrack$$
$$\mathbf{H}=\left\lbrack \begin{array}{ccc}
\mathbf{R} & \vdots  & \mathbf{p}\\
\cdots  & \cdot  & \cdots \\
\left\lbrack \begin{array}{ccc}
0 & 0 & 0
\end{array}\right\rbrack  & \vdots  & 1
\end{array}\right\rbrack$$
หากเรามีท่าทางทั้งทิศทางหมุน $\rotm RBG$ และตำแหน่ง $\pvec pGBG$ และทำการหมุนรอบแกน"ปัจจุบัน"ด้วยการหมุน $\mathbf{R}$แล้วทำการเลื่อนด้วยระยะ $\mathbf{p}$ เราจะได้ท่าทางใหม่เป็นดังต่อไปนี้
$$\left\lbrack \begin{array}{ccc}
{\mathbf{R}}^{\prime }  & \vdots  & {\mathbf{p}}^{\prime } \\
\cdots  & \cdot  & \cdots \\
\left\lbrack \begin{array}{ccc}
0 & 0 & 0
\end{array}\right\rbrack  & \vdots  & 1
\end{array}\right\rbrack =\left\lbrack \begin{array}{ccc}
\mathbf{R} & \vdots  & \mathbf{p}\\
\cdots  & \cdot  & \cdots \\
\left\lbrack \begin{array}{ccc}
0 & 0 & 0
\end{array}\right\rbrack  & \vdots  & 1
\end{array}\right\rbrack \left\lbrack \begin{array}{ccc}
\rotm RBG  & \vdots  & \pvec pGBG \\
\cdots  & \cdot  & \cdots \\
\left\lbrack \begin{array}{ccc}
0 & 0 & 0
\end{array}\right\rbrack  & \vdots  & 1
\end{array}\right\rbrack$$
หรือ
$${\mathbf{H}}^{\prime } =\rotm HBG$$
